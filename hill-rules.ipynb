```python
import string

def verify_symboles(name:str):
    """
    verifies if the name has symboles or is all numbers
    if retyrn true then pass jaweb berhi else refue straight up
    """
    if name.isdigit():
        return False  # all numbers
    for char in name:
        if not (char.isalnum() or char in [' ', '&']):
            return False  # unwanted symbol
    return True
```

```python
import google.generativeai as genai

# Configure your API key
genai.configure(api_key="AIzaSyC6yHwqS0J-5SZP7SNMoBxxfrGjK8a-5rk") # Replace with your actual API key

def get_user_input(prompt):
    """Helper function to get user input."""
    return input(prompt).strip()

def explain_company_types():
    """Briefly explains different company types in Tunisia."""
    print("\nIn Tunisia, company types include:")
    print("  - **SARL (Société à Responsabilité Limitée):** A limited liability company, suitable for small to medium-sized businesses. Owners' liability is limited to their capital contributions.")
    print("  - **SA (Société Anonyme):** A public limited company, typically for larger businesses with multiple shareholders. More complex regulatory requirements.")
    print("  - **SUARL (Société Unipersonnelle à Responsabilité Limitée):** A single-person limited liability company, similar to SARL but with a single owner.")
    print("  - **Société en nom collectif:** A general partnership where all partners have unlimited liability for the company's debts.")
    print("  - **Division of a foreign international company:** A branch or subsidiary of a company based outside Tunisia.")

def check_and_suggest_name(company_type, company_activity, proposed_name):
    """Checks the proposed name against legal rules and suggests corrections."""
    corrected_name = proposed_name
    needs_correction = False
    reasons = []

    # Rule 1: Division of a foreign company
    if company_type.lower() == "division of a foreign international company":
        if "فرع تونس" not in proposed_name and "division de Tunisie" not in proposed_name.lower():
            needs_correction = True
            reasons.append("For a division of a foreign company, the name must include 'فرع تونس' (Arabic) or 'division de Tunisie' (French).")
            if "فرع تونس" in proposed_name:
                corrected_name = proposed_name
            elif "division de Tunisie" in proposed_name.lower():
                corrected_name = proposed_name
            else:
                corrected_name_arabic = f"{proposed_name} فرع تونس"
                corrected_name_french = f"{proposed_name} division de Tunisie"
                print(f"**Suggestion:** You could use '{corrected_name_arabic}' or '{corrected_name_french}'.")
                return None, reasons, "Please choose one of the suggested forms or add the required phrase."

    # Rule 2: Société en nom collectif
    elif company_type.lower() == "société en nom collectif":
        if "و شركاهم" not in proposed_name and "et partenaires" not in proposed_name.lower():
            needs_correction = True
            reasons.append("For 'société en nom collectif', the name must include 'و شركاهم' (Arabic) or 'et partenaires' (French).")
            if "و شركاهم" in proposed_name:
                corrected_name = proposed_name
            elif "et partenaires" in proposed_name.lower():
                corrected_name = proposed_name
            else:
                corrected_name_arabic = f"{proposed_name} و شركاهم"
                corrected_name_french = f"{proposed_name} et partenaires"
                print(f"**Suggestion:** You could use '{corrected_name_arabic}' or '{corrected_name_french}'.")
                return None, reasons, "Please choose one of the suggested forms or add the required phrase."

    # Rule 3: Generic or misleading name (simple check, can be expanded)
    # This is a very basic check. A more robust solution would involve keyword analysis.
    generic_keywords = ["general", "company", "services", "solutions", "enterprise"]
    is_generic = any(keyword in proposed_name.lower() for keyword in generic_keywords)

    if is_generic and len(proposed_name.split()) <= 2: # Very simple heuristic for generic names
        needs_correction = True
        reasons.append("The proposed name appears too generic. It should be more specific to your company's activity.")
        # Propose more suitable name options (examples)
        print(f"**Suggestion:** Consider names that reflect '{company_activity}'. For example, if you are in 'IT consulting', you could use 'TechSolutions Consulting' or 'Innovative IT Partners'.")
        return None, reasons, "Please try to make your name more specific to your business."

    if needs_correction:
        return None, reasons, "Your proposed name needs adjustments."
    else:
        return proposed_name, [], "The name seems appropriate based on initial checks."

def main_conversation_flow():
    """Manages the conversation flow for company name generation."""
    print("Hello! I'm here to help you come up with an appropriate and legally valid company name in Tunisia.")

    # 1. Ask about company type
    company_type = get_user_input("\nWhat type of company are you planning to create? (SARL, SA, SUARL, société en nom collectif), or are you a division of a foreign international company? If you're unsure, type 'help'.")
    while company_type.lower() == 'help':
        explain_company_types()
        company_type = get_user_input("\nWhat type of company are you planning to create? (SARL, SA, SUARL, société en nom collectif), or are you a division of a foreign international company?")

    # 2. Ask about company activity
    company_activity = get_user_input("\nCan you tell me more about your company's activity or main business?")

    # 3. Ask about proposed name
    proposed_name = get_user_input("\nDo you already have a name in mind for your company?")

    # Check and suggest name
    final_name, reasons, message = check_and_suggest_name(company_type, company_activity, proposed_name)

    print("\n---")
    print("### Name Check Results:")
    if final_name:
        print(f"Your proposed name: **'{final_name}'** seems appropriate based on initial checks.")
    else:
        print(f"**Issue:** {message}")
        for reason in reasons:
            print(f"- {reason}")
        print("\nCould you please provide a revised name or consider the suggestions?")

    print("---\n")
    print("Remember, this is a preliminary check. You should always consult with a legal professional in Tunisia to ensure full compliance.")

# --- Main execution ---
```
